### Skewness Plot

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = ds.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
	plt.subplot(len(numerical_columns), 2, idx)
	sns.histplot(ds[feature], kde=True)
	plt.title(f"{feature} | Skewness: {round(ds[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

### Kurtosis Plot

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = ds.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
	plt.subplot(len(numerical_columns), 2, idx)
	sns.histplot(ds[feature], kde=True)
	plt.title(f"{feature} | Kurtosis: {round(ds[feature].kurt(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```
